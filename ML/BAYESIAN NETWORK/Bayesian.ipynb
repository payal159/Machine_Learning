```python
import pandas as pd
```

```python
# import bayesian network class(for discrete/categorical data)
from pgmpy.models import DiscreteBayesianNetwork
```

```python
# import estimator to learn probabilities from data
from pgmpy.estimators import MaximumLikelihoodEstimator
```

```python
# import inference method for prediction
from pgmpy.inference import VariableElimination
```

```python
# 2. load dataset from CSV file into a dataframe
data=pd.read_csv('/content/heart.csv')
data
```

```python
# 3. Create Bayesian Network structure (DAG)
# Each tuple represents a directed edge (parent → child)
model = DiscreteBayesianNetwork([
    ('age', 'target'),        # Age influences heart disease
    ('sex', 'target'),        # Gender influences heart disease
    ('cp', 'target'),         # Chest pain influences heart disease
    ('chol', 'target'),       # Cholesterol influences heart disease
    ('thalach', 'target')     # Heart rate influences heart disease
])
```

```python
# 4. Train the model using dataset
# It calculates conditional probability tables (CPTs)
model.fit(data, estimator=MaximumLikelihoodEstimator)
```

```python
# 5. Create inference object
# This is used to perform probability queries
inference = VariableElimination(model)
```

```python
# 6. Query the model to predict heart disease probability
# 'evidence' = given patient data
result = inference.query(
    variables=['target'],     # We want to predict 'target'
    evidence={
        'age': 56,           # Patient age
        'sex': 1,            # Male (1)
        'cp': 1,             # Chest pain type
        'chol': 250,         # Cholesterol level
        'thalach': 150       # Heart rate
    }
)
```

```python
# 7. Print the result
# Shows probability of having heart disease (1) or not (0)
print(result)
```
